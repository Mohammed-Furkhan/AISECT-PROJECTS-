```python
#importing Library
import pandas as pd
```

```python
#read Dataset
dataset=pd.read_csv("/content/Salary.csv")
```

```python
#Showing the dataset
dataset
```

```python
#Assigning the Independent(X) Values
#Assinging the Dependent(y) Values

x = dataset[["YearsExperience"]]
y = dataset[["Salary"]]
```

```python
#importing train_test_split from scikit
from sklearn.model_selection import train_test_split
```

```python
#splitting the train and test data's
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
#showing the x_test
x_test
```

```python
#from Skikit importing standardscalar
from sklearn.preprocessing import StandardScaler
```

```python
#Calling the StandardScaler
scaler = StandardScaler()
```

```python
#transformming the x data
scaler.fit_transform(x)
```

```python
#transformming the y data
scaler.fit_transform(y)
```

```python
#importing the LinearRegrsssion from scikit
from sklearn.linear_model import LinearRegression
```

```python
#assigning the regression
reg=LinearRegression()
```

```python
#training the model
reg.fit(x_train,y_train)
```

```python
#assiging the y_pred
y_pred = reg.predict(x_test)
```

```python
#print y_pred
y_pred
```

```python
#importing r_score from scikit learn
from sklearn.metrics import r2_score
```

```python
#finding the r2_score
r2_score(y_test,y_pred)
```
