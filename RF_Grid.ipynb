```python
#importing the Libraies
import numpy as np
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
dataset
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
# Create a parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],       # Number of trees
    'max_depth': [None, 5, 10, 20],       # Depth of trees
    'min_samples_split': [2, 5, 10],      # Min samples to split node
    'min_samples_leaf': [1, 2, 5],        # Min samples in leaf
    'bootstrap': [True, False]
}

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')

# fitting the model for grid search
grid.fit(X, y)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, grid_predictions)
```

```python
# print classification report
from sklearn.metrics import classification_report
clf_report = classification_report(y, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y,grid.predict_proba(X)[:,1])
```

```python
# Assuming 'grid' is your GridSearchCV object that has been fitted to the data
best_model = grid.best_estimator_
```

```python
import pickle
filename="RF_G.sav"
pickle.dump(best_model,open(filename,'wb'))
```
